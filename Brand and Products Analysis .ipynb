# Cleaning and Handling of Product Data 

```python
product.isnull().sum()
```

```python
product.loc[product.reviews.isnull()]
```

```python
product[(product['rating'].isna()) & (product['reviews'].isna())]
```

```python
## Reviews 
product.reviews.fillna(value = 0 , inplace = True)
```

```python
## For Ratings 
average_ratings = product.rating.mean

# Fill NaN values of ratings with the average rating of the respective product_name 
product['rating'].fillna(average_ratings, inplace = True)
```

```python
product["size"].isnull().sum()
```

```python
## Size 

product.groupby('product_name')['size'].unique().value_counts()

# Fill NaN values in 'size' column with '1.7 oz/ 50 mL'   ## max used size 
product['size'].fillna('1.7 oz/ 50 mL', inplace=True)  
```

```python
## variation_type

product.groupby('product_name')['variation_type'].unique().value_counts()

# Fill NaN values in 'size' column with 'Size+Concentration+ Formulation'  
product['variation_type'].fillna('Size+Concentration+ Formulation', inplace=True)  
```

```python
## variation_value

product.groupby('product_name')['variation_value'].unique().value_counts()

# Fill NaN values in 'size' column with 'Not Known'  
product.loc[:,'variation_value'].fillna('Not Known', inplace=True)  
```

```python
## Ingredients

product.groupby('product_name')['ingredients'].unique().value_counts()

# Fill NaN values in 'size' column with 'Not Known'  
product.loc[:,'ingredients'].fillna('Not Known', inplace=True)  
```

```python
product.head()   # I have dropped the variation_desc 
```

```python
## Highlights 

# Fill NaN values in 'highlights' column with 'Very good overall'
product.loc[:, 'highlights'].fillna('Very good overall', inplace=True)
```

```python
product.drop(['variation_desc', 'value_price_usd', 'sale_price_usd'], axis=1, inplace=True)
```

```python
## secondary_category
product.groupby('secondary_category')['secondary_category'].unique().value_counts()


# Fill NaN values in 'tertiary_category ' column with 'Very good overall'
product.loc[:, 'secondary_category'].fillna('Body', inplace=True)
```

```python
## tertiary_category

# Fill NaN values in 'tertiary_category ' column with 'Very good overall'
product.loc[:, 'tertiary_category'].fillna('Unknown', inplace=True)
```

```python
product.drop(['child_max_price', 'child_min_price'], axis=1, inplace=True)
```

# 4. Availability and Exclusivity of products

```python
limited_edition_percentage = (product['limited_edition'].sum() / len(product)) * 100
new_percentage = (product['new'].sum() / len(product)) * 100
online_only_percentage = (product['online_only'].sum() / len(product)) * 100
sephora_exclusive_percentage = (product['sephora_exclusive'].sum() / len(product)) * 100

print(f"Percentage of Limited Edition Products: {limited_edition_percentage:.2f}%")
print(f"Percentage of New Products: {new_percentage:.2f}%")
print(f"Percentage of Online Only Products: {online_only_percentage:.2f}%")
print(f"Percentage of Sephora Exclusive Products: {sephora_exclusive_percentage:.2f}%")  # for 2 digits after decimals
```

```python
attributes = ['Limited Edition', 'New', 'Online Only', 'Sephora Exclusive']
percentages = [limited_edition_percentage, new_percentage, online_only_percentage, sephora_exclusive_percentage]

plt.figure(figsize=(10, 6))
plt.bar(attributes, percentages, color=['blue', 'green', 'orange', 'purple'])
plt.title('Percentage of Products with Different Attributes')
plt.xlabel('Product Attributes')
plt.ylabel('Percentage (%)')
plt.ylim(0, 50)  
plt.tight_layout()

plt.show()
```

```python
## Comparision between  the 'loves_count' and 'rating' of out_of_stock & in_stock product.

# Calculate the average 'loves_count' and 'rating' for in-stock and out-of-stock products
in_stock_avg_loves = product.loc[product['out_of_stock'] == 0, 'loves_count'].mean()
out_of_stock_avg_loves = product.loc[product['out_of_stock'] == 1, 'loves_count'].mean()

in_stock_avg_rating = product.loc[product['out_of_stock'] == 0, 'rating'].mean()
out_of_stock_avg_rating = product.loc[product['out_of_stock'] == 1, 'rating'].mean()

# Bar plot to visualize the comparison

attributes = ['In Stock', 'Out of Stock']
avg_loves = [in_stock_avg_loves, out_of_stock_avg_loves]
avg_rating = [in_stock_avg_rating, out_of_stock_avg_rating]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot for 'loves_count'
axes[0].bar(attributes, avg_loves, color=['blue', 'orange'])
axes[0].set_title('Average Loves Count for In Stock vs Out of Stock')
axes[0].set_xlabel('Availability Status')
axes[0].set_ylabel('Average Loves Count')

# Plot for 'rating'
axes[1].bar(attributes, avg_rating, color=['green', 'purple'])
axes[1].set_title('Average Rating for In Stock vs Out of Stock')
axes[1].set_xlabel('Availability Status')
axes[1].set_ylabel('Average Rating')

plt.tight_layout()
plt.show()
```

```python
## Compare the 'loves_count' and 'rating' of new products with existing ones

# Separation of the data into new products and existing ones

new_products = product[product['new'] == 1]
existing_products = product[product['new'] == 0]

# Calculatiom of  the average 'loves_count' and 'rating' for new products and existing ones
avg_loves_count_new = new_products['loves_count'].mean()
avg_rating_new = new_products['rating'].mean()

avg_loves_count_existing = existing_products['loves_count'].mean()
avg_rating_existing = existing_products['rating'].mean()

# Using Subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Subplot for love_count
categories = ['New Products', 'Existing Products']
avg_loves_count = [avg_loves_count_new, avg_loves_count_existing]
ax1.bar(categories, avg_loves_count, color='blue')
ax1.set_title('Comparison of Average Loves Count')
ax1.set_xlabel('Product Type')
ax1.set_ylabel('Average Loves Count')

# Subplot for Rating 
avg_rating = [avg_rating_new, avg_rating_existing]
ax2.bar(categories, avg_rating, color='green')
ax2.set_title('Comparison of Average Rating')
ax2.set_xlabel('Product Type')
ax2.set_ylabel('Average Rating')

plt.tight_layout()


plt.show()
```

```python
## Investigating whether products with higher 'loves_count' tend to have higher 'rating' or more 'reviews'.
# Create a scatter plot between Loves Count and Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rating', y='loves_count', data=product, hue='reviews', size='reviews', sizes=(20, 200))
plt.title('Loves Count vs. Rating with Reviews')
plt.xlabel('Loves Count')
plt.ylabel('Rating')
plt.legend(title='Reviews')
plt.show()
```
